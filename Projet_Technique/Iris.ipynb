```python
#j'ai commencé l'algo d'apprentissage: Les tapes a suivre lors de l'application d'un algorithme
#j'ai divise l'ensemble de données en ensemble de données d'apprentissage et de test
```

```python
train, test = train_test_split(Iris, test_size = 0.3)
print(train.shape)
print(test.shape)
```

```python
train_X = train[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
train_y=train.Species
test_X= test[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']] 
test_y =test.Species  
```

```python
#L'ensemble de données de test est  plus petit que celui de la formation
#J'ai choisi l'algorithme de regression
#Passons a l'ensemble de données d'apprentissage a l'algorithme pour le former. 
#ici j'ai utilise la methode .fit ()
```

```python
#iport des librairies pour les methodes .fit() et .predict()
from sklearn import metrics 
from sklearn.tree import DecisionTreeClassifier
model = svm.SVC() #choisir l algorithme

model.fit(train_X,train_y) 
prediction=model.predict(test_X) 
print('The accuracy of the SVM is:',metrics.accuracy_score(prediction,test_y))
```

```python
from sklearn.linear_model import LogisticRegression 
model = LogisticRegression()
model.fit(train_X,train_y)
prediction=model.predict(test_X)
print('The accuracy of the Logistic Regression is',metrics.accuracy_score(prediction,test_y))
```

```python
#on passe ensuite les données de test a l'algorithme  pour prédire le résultat.
#ici j'ai utilisé la méthode .predict ().
```

```python
model=DecisionTreeClassifier()
model.fit(train_X,train_y)
prediction=model.predict(test_X)
print('The accuracy of the Decision Tree is',metrics.accuracy_score(prediction,test_y))
```

```python
from sklearn.neighbors import KNeighborsClassifier  

model=KNeighborsClassifier(n_neighbors=3) #ici on traite 3 points
model.fit(train_X,train_y)
prediction=model.predict(test_X)
print('The accuracy of the KNN is',metrics.accuracy_score(prediction,test_y))
```
